```python
import pandas as pd
import yfinance as yf
pd.set_option('mode.chained_assignment',None)
```

```python
lotsize = pd.read_csv("Downloads/fo_mktlots.csv")
bhavcopy = pd.read_csv("Downloads/cm22SEP2020bhav.csv")
bhavcopy = bhavcopy[bhavcopy['SERIES'] =='EQ']
bhavcopy['SYMBOL'] = bhavcopy['SYMBOL'].str.replace(' ','')
lotsize['SYMBOL    '] = lotsize['SYMBOL    '].str.replace(' ','')
lotsize = lotsize.iloc[3:,:5].reset_index(drop=True)
lotsize = lotsize.sort_values(by='SYMBOL    ').reset_index(drop=True)
mod_bhav = bhavcopy[bhavcopy['SYMBOL'].isin(lotsize['SYMBOL    '])].reset_index(drop=True)
lotsize['Close'] = mod_bhav['CLOSE'].astype('float')
lotsize['Lot Price'] = round(lotsize['Close']*lotsize['SEP-20     '].astype('int'),1)
lotsize = lotsize.sort_values(by='Lot Price').reset_index(drop=True)
ticker = list(mod_bhav['SYMBOL']+'.NS')
tick = list(mod_bhav['SYMBOL'])
```

```python
l = []
for i in range(len(ticker)):
    if ticker[i] != 'INFRATEL.NS':
        data = yf.download(tickers= ticker[i],period='2mo', interval='1mo')
        data2 = data.iloc[:1]
        data2['Symbol'] = ticker[i]
        data2['TodayClose'] = data['Close'][-1]
        data2['LotSize'] = int(lotsize[lotsize['SYMBOL    '] == tick[i]].iloc[0,2])
        l.append(data2)
    
monthly_data = pd.concat(l).reset_index(drop=True)   
l.clear()
```

```python
monthly_data['Symbol'] = monthly_data['Symbol'].str.replace('.NS','')
monthly_data['Margin'] = monthly_data['LotSize']*monthly_data['TodayClose']
monthly_data['Pivot'] = (monthly_data['Close']+monthly_data['High']+monthly_data['Low'])/3
monthly_data['S3'] = 2*monthly_data['Pivot'] - 2*monthly_data['High'] + monthly_data['Low']
monthly_data['S2'] = monthly_data['Pivot'] - monthly_data['High'] + monthly_data['Low']
monthly_data['L4'] = monthly_data['Close'] - (monthly_data['High']-monthly_data['Low'])*1.1/2
monthly_data['S4'] =  monthly_data['Pivot'] - 2*monthly_data['High'] + 2*monthly_data['Low']
monthly_data['L5'] =2*monthly_data['Close'] - monthly_data['High']*monthly_data['Close']/monthly_data['Low']
monthly_data = monthly_data.round(2)
```
